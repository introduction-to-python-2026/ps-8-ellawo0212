```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("sugar-consumption-by-country.csv")
country = 'Israel'
country_row = df[df['Country'] == country]
years = country_row.loc[:, '1990':].columns
values = country_row.loc[:, '1990':].iloc[0]
plt.plot(years, values)
plt.title(f'Sugar Consumption Over Time - Israel')
plt.xlabel('Year')
plt.ylabel('Grams per person per day')
plt.xticks(rotation=45)
plt.show()
```
